1. Esperanza matemática y su relación con la media de una distribución

La esperanza matemática (o valor esperado) de una variable aleatoria X es el promedio teórico de los valores que puede tomar, ponderado por sus probabilidades. Representa el valor promedio que se esperaría obtener si el experimento se repitiera un número muy grande de veces. En distribuciones de probabilidad, la esperanza coincide con la media poblacional; por ello, la esperanza se interpreta como la media de la distribución.


2. Definición de varianza usando el concepto de esperanza

La varianza mide qué tanto se dispersan los valores de la variable aleatoria con respecto a su media. Usando el concepto de esperanza, se define como:

Var(X) = E[(X - E[X])^2]

Es decir, la esperanza del cuadrado de las desviaciones respecto a la media. Otra forma equivalente es:

Var(X) = E[X^2] - (E[X])^2

3. Media y varianza de la distribución binomial

Si X ~ Binomial(n, p):

- Media: E[X] = n * p
- Varianza: Var(X) = n * p * (1 - p)


4. Identificación del modelo de la distribución Normal

La distribución Normal es una distribución continua que describe fenómenos donde los valores se concentran alrededor de una media y disminuyen simétricamente hacia ambos lados. Su función de densidad es:

f(x) = (1 / (σ * √(2π))) * exp(-(x - μ)^2 / (2σ^2))

Se denota como:
X ~ N(μ, σ^2)

5. Parámetros de la distribución normal

La distribución Normal está completamente definida por dos parámetros:

- μ : media (centro de la distribución)
- σ^2 : varianza (dispersión de los datos)

El parámetro σ es la desviación estándar.


6. Aproximación de la binomial mediante la distribución normal

Sí, la distribución binomial puede aproximarse mediante una distribución normal cuando el número de ensayos n es grande y la probabilidad p no es muy extrema. Una regla común es que la aproximación es adecuada cuando:

n * p ≥ 5  y  n * (1 - p) ≥ 5

En este caso, si X ~ Binomial(n, p), entonces puede aproximarse como:

X ≈ N(n * p, n * p * (1 - p))

En cálculos de probabilidades se recomienda aplicar la corrección de continuidad debido a que la binomial es discreta y la normal es continua.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Define parameters for the Normal distribution
mu = 0       # Mean
sigma = 1    # Standard deviation

# Generate a range of x values
x = np.linspace(((mu - 4)+2)*sigma, ((mu + 4)+2)*sigma, 100)  # Covers most of the distribution

# Calculate the PDF (Probability Density Function)
pdf = stats.norm.pdf(x, mu, sigma)

print(x)
print(pdf)
```

```python
# Create the plot
plt.stem(x, pdf)
plt.title(f'Normal Distribution (mu={mu}, sigma={sigma})')
plt.grid(True)
plt.show()
```
